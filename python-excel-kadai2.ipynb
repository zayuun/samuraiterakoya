```python
import pandas as pd
df = pd.DataFrame({
    '日付': ['2023-05-17', '2023-05-18', '2023-05-19', '2023-05-20', '2023-05-21'],
    '社員名': ['山田', '佐藤', '鈴木', '田中', '高橋'],
    '売上': ['100', '200', '150', '300', '250'],
    '部門': ['メーカー', '代理店', 'メーカー', '商社', '代理店']
})

# 売上列を数値型に変換する
df['売上'] = pd.to_numeric(df['売上'])

# 「売上」列の平均値を計算し、「平均売上」列を作成
df['平均売上'] = df['売上'].mean()

# 平均売上列を数値型に変換する
df['平均売上'] = pd.to_numeric(df['平均売上'])


#「業績ランク」列を作成し、売上の業績ランクを設定
def performance(rank):
    result = ' ';

    if rank >= df['平均売上'][0] + 50:
      result = 'A';
    elif rank >= df['平均売上'][0]:
      result = 'B';
    else:
      result = 'C';
    return result
df['業績ランク'] = df['売上'].apply(performance)

print(df['業績ランク'])

writer = pd.ExcelWriter('業績.xlsx')

# DataFrameオブジェクトをExcelファイルに書き込む
df.to_excel(writer, sheet_name='Sheet1', index=False)

# Excelファイルを保存する
writer.close()
```
